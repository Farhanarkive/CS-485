```python
import pandas as pd

df = pd.read_csv("train.csv")
```

```python
print(df.columns)
```

```python

# Gender columns
gender_cols = ["male", "female", "transgender", "other_gender"]
# Get groups: comments where the identity column > 0
gender_groups = {col: df[df[col] > 0] for col in gender_cols}
# Find the smallest group size
min_size = min(len(group) for group in gender_groups.values() if len(group) > 0)
print("Min group size:", min_size)
# Sample equally from each gender group
balanced_gender_df = pd.concat([group.sample(min_size, random_state=42)
                      for group in gender_groups.values() if len(group) > 0])
# Shuffle
balanced_gender_df = balanced_gender_df.sample(frac=1, random_state=42)
# Save
balanced_gender_df.to_csv("balanced_gender.csv", index=False)
print("Balanced gender dataset saved as balanced_gender.csv")
print("Final size:", len(balanced_gender_df))
```

```python

# Race identity columns
race_cols = ["black", "white", "asian", "latino", "other_race_or_ethnicity"]

# Group by identity > 0
race_groups = {col: df[df[col] > 0] for col in race_cols}

# Minimum available group size
min_size = min(len(group) for group in race_groups.values() if len(group) > 0)

print("Min group size:", min_size)

# Sample equally
balanced_race_df = pd.concat(
    [group.sample(min_size, random_state=42) for group in race_groups.values() if len(group) > 0]
)

# Shuffle rows
balanced_race_df = balanced_race_df.sample(frac=1, random_state=42)

# Save
balanced_race_df.to_csv("balanced_race.csv", index=False)

print("Balanced race dataset saved as balanced_race.csv")
print("Final size:", len(balanced_race_df))
```

```python
religion_cols = [
    "christian",
    "jewish",
    "muslim",
    "hindu",
    "buddhist",
    "atheist",
    "other_religion"
]

# Get groups
religion_groups = {
    col: df[df[col] > 0].copy()
    for col in religion_cols
    if df[df[col] > 0].shape[0] > 0
}

# Balance on smallest group
min_size = min(len(g) for g in religion_groups.values())

print("Balancing to size:", min_size)

# Equal sampling
balanced_rel_samples = [
    g.sample(min_size, random_state=42)
    for g in religion_groups.values()
]

balanced_religion_df = pd.concat(balanced_rel_samples)

# Shuffle
balanced_religion_df = balanced_religion_df.sample(frac=1, random_state=42)

# Save clean dataset
balanced_religion_df.to_csv("balanced_religion_dataset.csv", index=False)

print("Saved balanced religion dataset with shape:", balanced_religion_df.shape)
```

```python

# Sexuality-related identity columns in Jigsaw
sexuality_cols = [
    "heterosexual",
    "homosexual_gay_or_lesbian",
    "bisexual",
    "other_sexual_orientation"
]

# For each identity, take rows where the identity value > 0
sexuality_groups = {
    col: df[df[col] > 0].copy()
    for col in sexuality_cols
    if df[df[col] > 0].shape[0] > 0
}

# Find smallest group size to balance
min_size = min(len(g) for g in sexuality_groups.values())

print("Balancing to size:", min_size)

# Sample each group equally
balanced_samples = [
    g.sample(min_size, random_state=42)
    for g in sexuality_groups.values()
]

# Combine into one dataset
balanced_sexuality_df = pd.concat(balanced_samples)

# Shuffle
balanced_sexuality_df = balanced_sexuality_df.sample(frac=1, random_state=42)

# Save clean dataset WITH ONLY ORIGINAL COLUMNS
balanced_sexuality_df.to_csv("balanced_sexuality_dataset.csv", index=False)

print("Saved balanced sexuality dataset with shape:", balanced_sexuality_df.shape)
```
